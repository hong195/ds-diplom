## Импорт библиотек

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

## Feature engeeniring

```python
df_sessions = pd.read_csv('./data/train/df_sessions.csv', low_memory=False)

df_sessions
```

```python
categories = [col for col in df_sessions.columns if col.startswith('utm_') or col.startswith('device_') or col.startswith('geo_')]
```

```python
for category in categories:
    
    df_sessions[category] = df_sessions[category].astype('category')
    
df_sessions.dtypes
```

```python
def handle_keyword_len(keyword):
    return 0 if keyword == 'unknown' else len(keyword) 


df_sessions['utm_keyword_len'] = df_sessions['utm_keyword'].apply(handle_keyword_len)


df_sessions
```

```python
organic_trafic = ['organic', 'referral', '(none)']

df_sessions['is_organic_traffic'] = df_sessions.utm_medium.isin(organic_trafic)
```

```python
df_sessions['payable_trafic'] = df_sessions.is_organic_traffic.apply(lambda x: x != True)

df_sessions
```

```python
social_add = ['QxAxdyPLuQMEcrdZWdWb', 'MvfHsxITijuriZxsqZqt', 'ISrKoXQCxqqYvAZICvjs','IZEXUFLARCUMynmHNBGo', 'PlbkrSYoHuZBWfYjYnfw','gVRrcxiDQubJiljoTbGm']

df_sessions['is_social_add'] = df_sessions.utm_source.isin(social_add)

df_sessions
```

```python
def handle_screen_res(res):
    
    if (res == '(not set)'):
        return (0, 0)
    
    return res.split('x')

df_sessions['screen_res_width'] = df_sessions.device_screen_resolution.apply(lambda x: handle_screen_res(x)[0]).astype(int)
df_sessions['screen_res_height'] = df_sessions.device_screen_resolution.apply(lambda x: handle_screen_res(x)[1]).astype(int)

df_sessions
```

```python
ohe = OneHotEncoder(sparse_output=True)
ohe
```

```python
cats = ['utm_source',
 'utm_medium',
 'utm_campaign',
 'utm_keyword',
 'device_category',
 'device_os',
 'device_brand',
 'device_browser',
 'geo_country',
 'geo_city']
    
ohe.fit(df_sessions[cats])
```

```python
transformed_data = ohe.transform(df_sessions[cats])

transformed_data_names = ohe.get_feature_names_out()

transformed_df = pd.DataFrame.sparse.from_spmatrix(data=transformed_data, columns=transformed_data_names).astype('int8')

transformed_df
```

```python
std_scaler = StandardScaler()
```

```python
std_cols = ['utm_keyword_len', 'screen_res_width', 'screen_res_height']
std_count_vars = std_scaler.fit_transform(df_sessions[std_cols])
std_count_vars
```

```python
transformed_df[['utm_keyword_len_std', 'screen_res_width_std', 'screen_res_height_std']] = std_count_vars
transformed_df
```

```python
transformed_df['is_target_action'] = df_sessions['is_target_action'].apply(lambda x: 1 if x == True else 0)
transformed_df['is_organic_traffic'] = df_sessions['is_organic_traffic'].apply(lambda x: 1 if x == True else 0)
transformed_df['payable_trafic'] = df_sessions['payable_trafic'].apply(lambda x: 1 if x == True else 0)
transformed_df['is_social_add'] = df_sessions['is_social_add'].apply(lambda x: 1 if x == True else 0)

transformed_df
```

```python
transformed_df.to_pickle('./data/train/transformed.pickle')
```
